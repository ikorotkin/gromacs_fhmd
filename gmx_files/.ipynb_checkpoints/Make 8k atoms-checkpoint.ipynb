```python
f = open('approximately_8k_atoms_from_64k_atoms/conf64.gro', 'r')
text = f.read()
f.close()
```

```python
lines = text.splitlines()
lines[3:8]
```

```python
size = [float(d)/2 for d in lines[len(lines) - 1].split()]
print(size)
count = 0
coords_all = []

f = open('approximately_8k_atoms_from_64k_atoms/conf8.gro', 'w')
for k in range(len(lines)):
    if k == 0: 
        print(lines[k], file=f)
    elif k == 1:
        print('NaN', file=f)
    elif k == len(lines) - 1:
        print('  ' + ' '.join([str(float(d)/2) for d in lines[k].split()]), file=f)
    else:
        lines[k] = lines[k].replace(r'Ar\s*', 'Ar')
        coords = list(map(float, lines[k].split()[2:5]))
        is_in_region = all([coords[i] < size[i] for i in range(3)])
        if is_in_region:
            count += 1
            coords_all.append(coords)
            print(str(count) + 'Ar  Ar' + str(count) + ' '.join(lines[k].split()[1:8]), file=f)
    
    if k == 200:
        break
    
        
f.close()
print(len(coords_all))
```

```python
f = open('8k_atoms_s_const/energy.xvg')

text = f.read()

X = []

# parsing
for line in text.split('\n'):
    if line != '':
        if line[0] in '#@' or line == '':
            if line[0:3] == '@ s':
                print(line[4:])
            continue

        values = map(float, line.split())
        print(line)
        X.append(list(values))


df = pd.DataFrame(X, columns=['index', 'Potential', 'Kinetic En.', 'Total Energy', 'Temperature']).drop('index', axis=1)
df.plot(ls='-', marker='.', ylim=(-5000,35000))
```
