```python
def plot_at_step(n):
    plt.scatter(range(7989), df[df['Step']==n].values[0][1:-1])
    plt.suptitle('Step %d' % n)
    x1,x2,y1,y2 = plt.axis()
    plt.axis((x1,x2,10000, -10000))
    plt.show()

for i in range(4790, 4795):
    plot_at_step(i)

    
```
